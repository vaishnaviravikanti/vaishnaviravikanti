```python

# (1)Conduct linear search after taking a number from the user

from collections import Counter
print("Enter 20 Numbers: ")
arr = []
for i in range(20):
  arr.insert(i, int(input()))

def linearSearch(arr, n, x):
  

    # Going through array sequencially
    for i in range(0, n):
        if (arr[i] == x):
            return i
    return -1

n = len(arr)
x=int(input("Number to search:"))
result = linearSearch(arr, n, x)
if(result == -1):
    print("Element not found")
else:
    print("Element found at index: ", result)


# (2)Conduct Binary search after taking a number from the user


def binarySearch(arr, x1, low, high):

    # Repeat until the low and high meet each other
    while low <= high:

        mid = low + (high - low)//2

        if arr[mid] == x1:
            return mid

        elif arr[mid] < x1:
            low = mid + 1

        else:
            high = mid - 1

    return -1



x1=int(input("Number to binarysearch:"))
result = binarySearch(arr, x1, 0, len(arr)-1)

if result != -1:
    print("Element is present at index " + str(result))
else:
    print("Not found")


#(3)Conduct Bubble sort after taking a number from the user


def bubbleSort(arr):
    
  # accessing each array element
  for i in range(len(arr)):

    # comparing
    for j in range(0, len(arr) - i - 1):

      # compare two adjacent elements
      # change > to < to sort in descending order
      if arr[j] > arr[j + 1]:

        # swapping elements if elements are not in the intended order
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp




bubbleSort(arr)

print('Sorted Array in Ascending Order:')
print(arr)


# (5) Find mean, median and mode of the dataset.

#mean

summ = sum(arr)
mean = summ / n
print("The mean or average of all these numbers (", arr, ") is", str(mean))


#median

arr.sort()
if n % 2 == 0:
    median1 = arr[n//2]
    median2 = arr[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = arr[n//2]
print("The median of the given numbers  (", arr, ") is", str(median))


#mode

from collections import Counter
n= len(arr)
val = Counter(arr)
findMode = dict(val)
mode = [i for i, v in findMode.items() if v == max(list(val.values()))]  
if len(mode) == n:
    findMode = "The group of number do not have any mode"
else:
    findMode = "The mode of a number is / are: " + ', '.join(map(str, mode))
print(findMode)


#(6)Find all prime numbers in the list.

prime=[]
for i in arr:
    c=0
    for j in range(1,i):
        if i%j==0:
           c+=1
    if c==1:
       prime.append(i)
print(prime)
```

DEFINATION:

Bubble sort is a sorting algorithm that compares two adjacent elements and swaps them until they are in the intended order.

PROCEDURE:

Comparing the current element with the one after it, swapping their values if needed. These passes through the list are repeated until no swaps had to be performed during a pass, means that until list has become fully sorted. The algorithm, is also called as comparison sort.

This sorting is performed from left to right, switching their positions if they are out of order.
